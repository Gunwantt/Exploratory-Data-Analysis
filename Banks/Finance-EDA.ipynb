# EDA

```python
df.head()
```

```python
banks = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

### Max Close Price for each bank's Stock

```python
df.xs(key='Close', axis=1, level='Stock Info').max()
```

### Creating dataframe for returns on each banks stock

```python
returns_df = pd.DataFrame()
```

```python
for bank in banks:
    returns_df[bank + ' Return'] = df[bank]['Close'].pct_change()
```

```python
returns_df.head()
```

#### 1st row is null as there is nothing in the past to compare it to 

```python
sns.pairplot(returns_df)
```

City Group had huge crash back in 2008, Hence the returns of CityGroup are unlike other stocks

### Checking the Best and Worst return date for each bank

```python
returns_df.head()
```

```python
returns_df.idxmin()
```

#### 4 of the banks share same date for worst drop

```python
returns_df.idxmax()
```

### Checking which stock was the riskiest across entire period

```python
returns_df.std().sort_values(ascending=False)
```

C Returns was the riskiest of all because of very high standard deviation compared to other stocks

### Checking fluctuations in the year 2015

```python
returns_df.loc['2015-01-01':'2015-12-31',:].std().sort_values(ascending=False)
```

```python
sns.set_style('darkgrid')
```

```python
sns.distplot(returns_df.loc['2015-01-01':'2015-12-31','MS Return'], bins=40)
```

Returns on MS were very stable with deflection of about 0.06 only

```python
sns.distplot(returns_df.loc['2008-01-01':'2008-12-31','C Return'], bins=40, color='g')
```

In this case, the deflection is very high which is 0.2

```python
plt.figure(figsize=(7,6))
mask = np.triu(df.xs(key='Close', level='Stock Info', axis=1).corr())
sns.heatmap(close_corr, annot=True, cmap='coolwarm', mask=mask)
```
